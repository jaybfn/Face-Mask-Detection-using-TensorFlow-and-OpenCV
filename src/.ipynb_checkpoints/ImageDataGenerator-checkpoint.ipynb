```python
#IDG = ImageDataGenerator(brightness_range=[0.2,1.0])

def generate_new_images(IDG, file_name_prefix):
    files_in_path = os.listdir(save_path)  
    
    for class_ in classes:
        files = os.listdir(base_path + '/' + class_)
        
        if len(files_in_path) == 0:
            img_path = os.path.join(save_path, class_)
            os.mkdir(img_path)
            
            for file in files:
                img = load_img(path=base_path + '/' + class_+'/'+f'{file}',target_size=(224,224))
                x = np.array(img)
                sample = np.expand_dims(x,axis = 0)
                iterator = IDG.flow(sample, batch_size=1)
                print(iterator)

                for i in range(9):
                    plt.subplot(330 +1 +i)
                    batch = iterator.next()
                    img = batch[0].astype('uint8')
                    save_img(img_path+'\\'+ f'{file_name_prefix}_{i}_{file}', img)
                    plt.imshow(img)
                plt.show()
        else:
            img_path = os.path.join(save_path, class_)
            for file in files:
                img = load_img(path=base_path + '/' + class_+'/'+f'{file}',target_size=(224,224))
                x = np.array(img)
                sample = np.expand_dims(x,axis = 0)
                iterator = IDG.flow(sample, batch_size=1)
                print(iterator)

                for i in range(9):
                    plt.subplot(330 +1 +i)
                    batch = iterator.next()
                    img = batch[0].astype('uint8')
                    save_img(img_path+'\\'+ f'{file_name_prefix}_{i}_{file}', img)
                    plt.imshow(img)
                plt.show()

           
```
