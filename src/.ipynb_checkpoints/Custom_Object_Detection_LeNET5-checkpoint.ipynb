```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.utils import to_categorical
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# from google.colab import drive
# drive.mount('/content/gdrive')
```

```python
# Let's explore the data folder
base_path = r"../data_face/SyntheticImageGenerator"
os.listdir(base_path)   
```

```python
# Let's define the classes
classes = os.listdir(base_path)
```

```python
X_list = []
y_list = []
for class_ in classes:
    files = os.listdir(base_path + '/' + class_)
    #print(files)
    for file in files:
        img = load_img(path=base_path + '/' + class_+'/'+f'{file}',target_size=(224,224))
        x = np.array(img)
        X_list.append(x)
        y_list.append(class_)
        
X = np.array(X_list)/255
y = np.array(y_list)
```

```python
#plot the image with labels
plt.figure(figsize=(16,16))
for i in range(51):
    plt.subplot(8,8,i+1,title=f'Class:{y[i]}')
    plt.imshow(X = X[i])
    plt.axis('off')
```

```python
y_series = pd.Series(y).map({classes[0]:0, classes[1]:1})
y_series
```

```python
y = to_categorical(y_series)
```

```python
np.argmax(y,axis = 1)
```

```python
from tensorflow import keras
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout
from tensorflow.keras.activations import relu, softmax
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as K
import tensorflow as tf
from keras.regularizers import l2
```

```python
x_train, x_test, y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state=42)
```

```python
%%time
K.clear_session()
model = Sequential([
### Implement LeNet-5 here
    
    # Conv1 with relu activation function # input layer
    Conv2D(filters= 6, kernel_size=(5,5), strides=(1,1), input_shape=x_train[0].shape,
           activation=keras.activations.relu,
           padding='valid', kernel_regularizer=l2(0.0001)), # valid means no padding

    # Pool1 with no padding
    MaxPooling2D(pool_size=(2,2), strides=(2,2), padding= 'valid'),
    
    # Conv2 with relu activation function
    Conv2D(filters = 16, kernel_size=(5,5),strides=(1,1),
           activation=keras.activations.relu, 
           padding = 'valid',kernel_regularizer=l2(0.0001)), # valid means no padding
    
    # Pool2 with no padding
    MaxPooling2D(pool_size=(2,2), strides=(2,2), padding= 'valid'), # valid means no padding
    
    
    # Flatten layer
    Flatten(),
    
    # Fully connected dense layer FC3 with relu activation function
    Dropout(0.2),
    Dense(units=256, activation=keras.activations.relu),
    BatchNormalization(),
    

    # Fully connected dense layer FC4 with relu activation function
    Dropout(0.2),
    Dense(units=128, activation=keras.activations.relu),
    BatchNormalization(),

    Dropout(0.2),
    Dense(units=64, activation=keras.activations.relu),
    BatchNormalization(),

    Dropout(0.2),
    Dense(units=32, activation=keras.activations.relu),
    BatchNormalization(),

    # Fully connected output layer with softmax
    Dense(units=2, activation=keras.activations.softmax)   # How many neurons? we want classify two classes
    
])
```

```python
model.compile(optimizer=keras.optimizers.Adam(0.5e-4), loss=keras.losses.categorical_crossentropy, metrics=['accuracy'])
```

```python
cb = keras.callbacks.EarlyStopping(
    monitor="accuracy",
    min_delta=0.0005,
    patience=10,
    verbose=1,
    mode="min",
    baseline=None,
    restore_best_weights=False,)
```

```python
%%time
history = model.fit(x_train,y_train, 
                    epochs = 100, 
                    batch_size = 10, 
                    verbose = 1, 
                    validation_split=0.3,
                    callbacks=[cb]) #callbacks=[cb] #30

# 20, 20
#100,200
#100,150
#50,50
# 100, 50 , lr = 1e-4
# 100, 30 , lr = 1e-4 --good
# 100, 30 , lr =0.5e-4 --best
```

```python
pd.DataFrame(history.history).plot()
plt.grid(True)
plt.gca().set_ylim(0,1.1) # set the y range to [0,1]
```

```python
model.evaluate(x_test, y_test)
```
